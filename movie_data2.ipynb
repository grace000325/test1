```python
def get_grade(name):
  #네이버 영화 검색 페이지
  url = 'https://search.naver.com/search.naver?sm=tab_sug.top&where=nexearch&query={}'
  r = requests.get(url.format(name))
  soup = BeautifulSoup(r.text,'html.parser')
  movie_info = soup.find('a',attrs={'class':'more_link'})
  if movie_info == None:
    return ''
  #상세페이지 가져오기
  url = 'https://search.naver.com/search.naver'+movie_info['href']
  r = requests.get(url)
  soup = BeautifulSoup(r.text,'html.parser')
  movie_detail = soup.find('dl',attrs={'class':'info txt_4'})
  movie_div_list=movie_detail.find_all('div')
  for div in movie_div_list:
    if div.find('dt').get_text() == '등급':
      grade_dd = div.find('dd').get_text()
      return grade_dd
  return 1
```

```python
get_grade('가디언즈 오브 갤럭시: Volume 3')
```

```python
def get_grade2(row):
  name = row['movieNm']
  return get_grade(name)
```

```python
current_top5_movie = movies.head()
current_top5_movie
```

```python
grade = current_top5_movie.apply(get_grade2, axis=1)
grade.name = 'grade'
```

```python
current_top5_movie_with_grade = pd.merge(current_top5_movie,grade, left_index=True, right_index=True)
current_top5_movie_with_grade
```

```python
#일주일간 범죄도시3의 해당일매출액(salesAmt), 해당일관객수(audiCnt) 변화 선그래프 
```

```python
date = []
for i in range(20230601, 20230608):
  date.append(str(i))

weekmovie = pd.DataFrame()

for i in date:
  url = "http://kobis.or.kr/kobisopenapi/webservice/rest/boxoffice/searchDailyBoxOfficeList.json?key=2b32a504ab7d7ee63f1635f16f7a00f0&targetDt=" + i
  res = requests.get(url)
  text= res.text

  data = json.loads(text)
  movies = []
  for d in data['boxOfficeResult']['dailyBoxOfficeList']:
    movies.append(d)
  movies = pd.DataFrame(movies)
  movie = movies[movies['movieNm'] == '범죄도시3']

  grade = movie.apply(get_grade2, axis=1)
  grade.name = 'grade'
  movie = pd.merge(movie,grade, left_index=True, right_index=True)

  weekmovie = pd.concat([weekmovie,movie], ignore_index=True)

weekmovie
```

```python
import matplotlib.pyplot as plt

day = ['0601', '0602', '0603', '0604', '0605', '0606', '0607']
audiCnt = weekmovie.audiCnt

audiCnt = [int(i) for i in audiCnt]
days = [int(i) for i in day]

fig, ax = plt.subplots()

plt.rcParams['figure.dpi']=100

ax.plot(day, audiCnt, '.:r')
# ax.plot(day, salesAmt, '.:g')
#척도 조정: 값의 차이가 큰 경우, 척도(scale)를 조정하여 그래프를 그림. 값을 로그 스케일이나 지수 스케일로 변환하여 작은 값의 변화를 더 잘 볼 수 있도록 함
# plt.yscale('log')

current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:.0f}'.format(x) for x in current_values])

# for i in range(1,7):
#   ax.annotate(audiCnt[i],(audiCnt[i],days[i]),xytext=(audiCnt[i]+0.2, days[i]))

for i, (x,y) in enumerate(zip(day,audiCnt)):
  plt.annotate(f'{y}', (x, y), xytext=(5, 5),
                 textcoords='offset points', ha='center', fontsize=8,
                 arrowprops=dict(arrowstyle='->'))

ax.set_title('한 주의 범죄도시 3의 관객 수')
ax.set_xlabel('day')
ax.set_ylabel('number')
plt.show()
```
